```python
import os, json
import numpy as np

meta_file = 'metadata - old DeepSTABp-lysates dataset (test set), available from dp180.csv'
accession_file = 'accessions - variants as separate proteins - cell_lysate-0_1 - 20241122 - all files - test.txt'
```

```python
metadata_raw = np.loadtxt(
    meta_file,
    delimiter=',',
    dtype=np.str_
)
assert metadata_raw.shape[0] == np.unique(metadata_raw[:, 0]).size

accessions_of_interest = np.loadtxt(
    accession_file,
    delimiter=',',
    dtype=np.str_
)
assert accessions_of_interest.size == np.unique(accessions_of_interest).size
print('# accessions of interest:', accessions_of_interest.size)

# find all accessions available in metadata_raw
accessions = np.array([
    acc for acc in accessions_of_interest
    if acc in metadata_raw[:, 0]
])
print('# accessions of interest present in metadata:', accessions.size)

metadata = np.concatenate(
    [
        metadata_raw[metadata_raw[:, 0] == acc]
        for acc in accessions
    ]
)

# print(metadata.shape)
```

```python
[metadata[metadata[:, 0] == acc] for acc in accessions]
```

```python
np.savetxt(
    'metadata - old DeepSTABp-lysates dataset (test set), available from dp180 - lysate_cell-0_1.csv',
    metadata,
    delimiter=',',
    fmt='%s'
)
```
