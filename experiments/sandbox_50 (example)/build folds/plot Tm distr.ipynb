```python
import numpy as np
import matplotlib.pyplot as plt

metafile = 'metadata - sandbox_50 (example).csv'
```

```python
metadata = np.loadtxt(
    metafile,
    delimiter=',',
    dtype=np.str_
)

print(metadata.shape)
```

```python
fig, ax = plt.subplots(
    figsize=(3.74, 3.74),
    dpi=300,
    layout='compressed'
)

ax.hist(
    metadata[:, 1].astype(np.float_),
    bins=25
)
ax.set_xlabel('Tm')
ax.set_ylabel('occurrences')

plt.savefig('histogram - Tm.png')
plt.show()
plt.close()
```
